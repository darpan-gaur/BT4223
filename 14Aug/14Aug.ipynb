```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
class Dynamics:
    def __init__(self):
        self.dim = 3
        self.num_particles = 500
        self.box_len = 10
        self.mass = 1
        self.temp = 300
        self.k = 1
        self.positions = np.zeros((self.num_particles, self.dim))
        self.velocities = np.zeros((self.num_particles, self.dim))
    
    def init_positions(self):
        self.positions = np.random.random_sample((self.num_particles, self.dim))*self.box_len
    
    def init_velocities(self):
        std_dev = (self.mass/(self.k*self.temp))**0.5
        self.velocities = np.random.normal(loc=0, scale=std_dev, size = (self.num_particles, self.dim))

    def draw_particles(self):
        plt.figure(figsize=(5, 5))
        axis = plt.gca()

        plt.title('Particles in 2D space')

        axis.set_xlim(-10, self.box_len+10)
        axis.set_ylim(-10, self.box_len+10)

        for i in range(self.num_particles):
            axis.add_patch(plt.Circle(self.positions[i], radius=0.5, linewidth=2, edgecolor='black'))
```

```python
dyn = Dynamics()
dyn.init_positions()
dyn.init_velocities()
```

```python
dyn.init_positions()
dyn.init_velocities()
dyn.draw_particles()
```

```python
# write positions in xyz file
with open('positions.xyz', 'w') as f:
    f.write(f'{dyn.num_particles}\n')
    f.write('Positions of particles\n')
    for i in range(dyn.num_particles):
        f.write(f'Ar ')
        for j in range(dyn.dim):
            f.write(f'{dyn.positions[i][j]} ')
        f.write('\n')

# write velocities in xyz file
with open('velocities.xyz', 'w') as f:
    f.write(f'{dyn.num_particles}\n')
    f.write('Velocities of particles\n')
    for i in range(dyn.num_particles):
        # f.write(f'Ar {dyn.velocities[i][0]} {dyn.velocities[i][1]} 0.0\n')
        f.write(f'Ar ')
        for j in range(dyn.dim):
            f.write(f'{dyn.velocities[i][j]} ')
        f.write('\n')
```
